```python
import pandas as pd

df = pd.read_csv("Mental Health Dataset.csv")

df.head()
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df['mental_health'] = df['treatment'].str.lower().map({'yes':1, 'no':0})
```

```python
df[df['mental_health'] == 0].head()
```

```python
cols = [
    'growing_stress',
    'changes_habits',
    'mental_health_history',
    'mood_swings',
    'coping_struggles',
    'work_interest',
    'social_weakness'
]

for col in cols:
    df[col] = df[col].str.strip().str.lower()
    df[col] = df[col].map({'yes':1, 'no':0})
```

```python
df.drop(['timestamp'], axis=1, inplace=True)
```

```python
from scipy import stats
```

```python
def run_ttest(feature):
    group1 = df[df[feature] == 1]['mental_health']
    group2 = df[df[feature] == 0]['mental_health']

    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

    print(f"{feature}:")
    print("T-stat =", t_stat)
    print("P-value =", p_value)

    if p_value < 0.05:
        print("Result: Significant\n")
    else:
        print("Result: Not Significant\n")
```

```python
features = [
    'growing_stress',
    'coping_struggles',
    'mental_health_history',
    'social_weakness',
    'changes_habits'
]

for f in features:
    run_ttest(f)
```

```python
impact = {}

for col in features:
    impact[col] = df.groupby(col)['mental_health'].mean()[1]

sorted(impact.items(), key=lambda x: x[1], reverse=True)
```
